# TP1 : Contrôle optimal


$$
\inf_{(x_1, x_2) \in K} x_1^2 + x_2^2 - 14x_1 - 6x_2 - 7 \\
\Longleftrightarrow \inf_{(x_1, x_2) \in K} J(x_1, x_2)
$$

avec $K = \{ (x_1, x_2) \in \mathbb{R}^2 | x_1 + x_2 \leq 2 \ \text{et} \ 2x_2 \leq 3\} $

Résolution théorique :

Les contraintes
$ g_1(x) = x_1 + x_2 - 2 \leq 0,\ g_2(x) = 2x_2 - 3 \leq 0 $
sont affines donc qualifiées.

Donc d'apres KKT, $\exists \mu_1, \mu_2 \geq 0$ tels que

$$
\begin{cases}
 \nabla J(x) + \mu_1 \nabla g_1(x) + \mu_2 \nabla g_2(x) = 0 \\
 \mu_1 g_1(x) = 0 \\
 \mu_2 g_2(x) = 0 \\
 g_1(x) \leq 0 \\
 g_2(x) \leq 0 \\
 \mu_1 \geq 0, \mu_2 \geq 0 \ non\ tous \ nuls
\end{cases}
$$

On a donc

$$
\begin{cases}
 2x_1 - 14 + \mu_1 + \mu_2 = 0 \\
 2x_2 - 6 + \mu_1 + 2\mu_2 = 0 \\
 \mu_1(x_1 + x_2 - 2) = 0 \\
 \mu_2(2x_2 - 3) = 0 \\
 \mu_1 \geq 0, \mu_2 \geq 0 \ non\ tous \ nuls
\end{cases}
$$

On a donc 4 cas à traiter :

- Cas 0: $\mu_1 = 0, \mu_2 = 0$ Impossible car $\mu_1 \mu_2 $ doivent être non tous les deux nuls
- Cas 1: $\mu_1 \neq 0, \mu_2 = 0$
  $$
  \begin{cases}
  2x_1 - 14 + \mu_1 = 0 \\
  2x_2 - 6 + \mu_1 = 0 \\
  x_1 + x_2 - 2 = 0 \\
  \mu_1 \geq 0
  \end{cases}
  $$

$$
\begin{cases}
x_1 = 7 - \frac{\mu_1}{2} \\
x_2 = 3 - \frac{\mu_1}{2} \\
\end{cases}
$$

$$
x_1 + x_2 - 2 = 0 \Rightarrow 10 - \mu_1 = 2 \Rightarrow \mu_1 = 8 \\
2x_1 - 14 + 8 = 0 \Rightarrow x_1 = 3 \\
x_1 + x_2 - 2 = 0 \Rightarrow x_2 = -1
$$

Donc $(3, -1, 8, 0)$ est un point candidat

- Cas 2: $\mu_1 = 0, \mu_2 \neq 0$

$$
\begin{cases}
2x_1 - 14 + \mu_2 = 0 \\
2x_2 - 6 + 2\mu_2 = 0 \\
2x_2 - 3 = 0 \\
\mu_2 \geq 0
\end{cases}
$$

$$
\begin{cases}
x_1 = 7 - \frac{\mu_2}{2} \\
x_2 = 3 - \mu_2 \\
2x_2 -3 = 0 \Rightarrow x_2 = \frac{3}{2}
\end{cases}
$$

$$
2x_2 - 6 + 2\mu_2 = 0 \Rightarrow 3 - 6 + 2\mu_2 = 0 \Rightarrow \mu_2 = \frac{3}{2}
$$

$$
2x_1 = 14 - \mu_2 \Rightarrow x_1 = 7 - \frac{3}{4} = \frac{25}{4}
$$

$$
2x_2 = 6 - 2\mu_2 = 6 - 3 = 3 \Rightarrow x_2 = \frac{3}{2}
$$

Donc $(\frac{25}{4}, \frac{3}{2}, 0, \frac{3}{2})$ pourrait être un point candidat, mais il n'est pas dans $K$ car il ne respecte pas la contrainte $x_1 + x_2 - 2 \leq 0$.  
En effet, $\frac{25}{4} + \frac{3}{2} - 2 = \frac{25}{4} + \frac{6}{4} - \frac{8}{4} = \frac{23}{4} > 0$

- Cas 3: $\mu_1 \neq 0, \mu_2 \neq 0$

$$
\begin{cases}
2x_1 - 14 + \mu_1 + \mu_2 = 0 \\
2x_2 - 6 + \mu_1 + 2\mu_2 = 0 \\
x_1 + x_2 - 2 = 0 \\
2x_2 - 3 = 0 \\
\mu_1 \geq 0, \mu_2 \geq 0
\end{cases}
$$

$$
\begin{cases}
x_1 = 7 - \frac{\mu_1 + \mu_2}{2} \\
x_2 = 3 - \frac{\mu_1}{2} - \mu_2 \\
\end{cases}
$$

$$
2x_2 - 3 = 0 \Rightarrow x_2 = \frac{3}{2}
$$

$$
x_1 + x_2 - 2 = 0 \Rightarrow x_1 = \frac{1}{2}
$$

$$
2x_1 - 14 + \mu_1 + \mu_2 = 0 \Rightarrow 2(\frac{1}{2}) - 14 + \mu_1 + \mu_2 = 0 \Rightarrow \mu_1 + \mu_2 = 13
$$

$$
2x_2 - 6 + \mu_1 + 2\mu_2 = 0 \Rightarrow 2(\frac{3}{2}) - 6 + \mu_1 + 2\mu_2 = 0 \Rightarrow \mu_1 + 2\mu_2 = 3
$$

$$
\begin{cases}
\mu_1 + \mu_2 = 13 \\
\mu_1 + 2\mu_2 = 3
\end{cases}
$$

$$
\begin{cases}
\mu_2 = -10 \\
\mu_1 = 23
\end{cases}
$$

Ce qui est impossible car $\mu_1 \geq 0, \mu_2 \geq 0$


Donc le seul point candidat est $(3, -1, 8, 0)$ et $J(3, -1) = -33$


```python
import numpy as np
import matplotlib.pyplot as plt
from gekko import GEKKO
```

```python
def J(x):
    return x[0] ** 2 + x[1] ** 2 - 14 * x[0] - 6 * x[1] - 7


print(J([3, -1]))
```

```python
x = np.linspace(-10, 20, 100)
y = np.linspace(-10, 20, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2 - 14 * X - 6 * Y - 7

fig = plt.figure()
ax = plt.axes(projection="3d")
ax.contour3D(X, Y, Z, 50)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title(r"Contour plot of f(x, y) = $x^2 + y^2 - 14x - 6y - 7$")
plt.show()
```

```python
# solve numerically with gekko
m = GEKKO(remote=True)
eq = m.Var(value=40)

x1 = m.Var(value=0, lb=-10, ub=20)
x2 = m.Var(value=0, lb=-10, ub=20)

m.Equation(x1 + x2 <= 2)
m.Equation(x1 + 2 * x2 <= 3)

m.Obj(x1**2 + x2**2 - 14 * x1 - 6 * x2 - 7)

m.options.IMODE = 3

m.solve(disp=False)

x = np.round(x1.value[-1], 2)
y = np.round(x2.value[-1], 2)

print("(x, y) = ({}, {})".format(x, y))
```

## Problème en temps minimal

### tram v1


```python
m = GEKKO()
nt = 501
m.time = np.linspace(0, 1, nt)

# Variables
x = m.Var(value=0)
y = m.Var(value=0)
p = np.zeros(nt)
p[-1] = 1
final = m.Param(value=p)

# Opitimize final time
tf = m.FV(value=1.0, lb=0.1, ub=100.0)
tf.STATUS = 1

# Control
u = m.MV(value=0, lb=-1, ub=1)  # |u(t)| <= 1
u.STATUS = 1

# Equations
m.Equation(x.dt() == y * tf)  # dx/dt = y
m.Equation(y.dt() == u * tf)  # dy/dt = u

# Terminal constraints
m.Equation(x * final == 0)  # x(1) = 0

# Terminal conditions as penalties in the objective function
terminal_penalty_y = 1e10 * (y * final + 1) ** 2  # Large weight on y(T) = -1

# Objective function
m.Obj(tf + terminal_penalty_y)

# Solve
m.options.IMODE = 6
m.solve(disp=False)

print("Final Time:", tf.value[0])
print("x(T):", x.value[-1])
print("y(T):", y.value[-1])

tm = np.linspace(0, tf.value[0], nt)

plt.figure(figsize=(12, 6))

plt.plot(tm, x.value, "k-", label=r"$x$ : tram speed")
plt.plot(tm, y.value, "b-", label=r"$y$ : tram acceleration")
plt.plot(tm, u.value, "r--", label=r"$u$ : control input")

plt.title("Optimal Control of a Tram to Minimize Time to Stop")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend(loc="best")

plt.tight_layout()
plt.show()
```

### tram v2


```python
m = GEKKO()
nt = 500
m.time = np.linspace(0, 1, nt)

epsilon = 0.1

# Variables
x = m.Var(value=0)
y = m.Var(value=0)
z = m.Var(value=0)
p = np.zeros(nt)
p[-1] = 1
final = m.Param(value=p)
alpha = m.Var(value=0.5, lb=0, ub=1)  # convex parameter


# Optimize final time
tf = m.FV(value=1.0, lb=0.1, ub=100.0)
tf.STATUS = 1

# Control
u = m.MV(value=0, lb=-1, ub=1)  # |u(t)| <= 1
u.STATUS = 1

# Equations
m.Equation(x.dt() == y * tf)  # dx/dt = y
m.Equation(y.dt() == u * tf)  # dy/dt = u
m.Equation(z.dt() == 0.5 * ((x * tf) ** 2 + (y * tf) ** 2))

# Terminal constraints
m.Equation(x * final == 0)  # x(T) = 0

# Penalize if y(T) < -1 - epsilon
penalty_lower = m.max2(0, (-1 - epsilon) - (y * final))
# Penalize if y(T) > -1 + epsilon
penalty_upper = m.max2(0, (y * final) - (-1 + epsilon))
penalty = 1e10 * (penalty_lower + penalty_upper)


# Objective function: minimize a convex combination of final time and integral of (ẋ(t)^2 + x(t)^2)
m.Obj(alpha * tf + (1 - alpha) * z * final + penalty)

# Solve
m.options.IMODE = 6
m.solve(disp=False)

print("Final Time: ", tf.value[0])
print("x(T): ", x.value[-1])
print("y(T): ", y.value[-1])
print("alpha: ", alpha.value[0])
print("penalty_lower: ", penalty_lower[-1])
print("penalty_upper: ", penalty_upper[-1])


tm = np.linspace(0, tf.value[0], nt)

plt.figure(figsize=(12, 6))

plt.plot(tm, x.value, "k-", label=r"$x$ : tram speed")
plt.plot(tm, y.value, "b-", label=r"$y$ : tram acceleration")
plt.plot(tm, u.value, "r--", label=r"$u$ : control input")
plt.plot(tm, alpha.value[0] * np.ones(nt), "g--", label=r"$\alpha$")
plt.title(
    "Optimal Control of a Tram to Minimize Time to Stop with Terminal Constraints"
)
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend(loc="best")

plt.tight_layout()
plt.show()
```

## Contrôle d'insectes


1. Montrons que $x(t) > 0$, $y(t) > 0$ pour tout $t \in [0, T]$.

$$
x' = x(1-y) \Rightarrow  \frac{x'}{x} = 1-y \\
\Leftrightarrow  \int_0^t \frac{x'(s)}{x(s)}ds = \int_0^t (1-y(s))ds \\
\Leftrightarrow ln(x(t)) - ln(\underbrace{x(0)}_{=1}) = \int_0^t (1-y(s))ds \\
\Leftrightarrow  x(t) = e^t e^{-\int_0^t y(s)ds}
$$
et 
$$
y' = -y(u - x) \Rightarrow  \frac{y'}{y} = -u + x \\
\Leftrightarrow  \int_0^t \frac{y'(s)}{y(s)}ds = \int_0^t (-u + x(s))ds \\
\Leftrightarrow ln(y(t)) - ln(\underbrace{y(0)}_{=4}) = \int_0^t (-u + x(s))ds \\
\Leftrightarrow  y(t) = 4e^{\int_0^t (-u + x(s))ds }
$$


$t \in [0,T]$, $u$ est borné, donc il suffit que $\int_0^t y(s)ds < \infty$ et $\int_0^t x(s)ds < \infty$ pour que $x(t) > 0$, $y(t) > 0$ pour tout $t \in [0, T]$ ce qui est le cas car $x(t)$ et $y(t)$ se controlent mutuellement.


2. Points d'équilibre

On cherche $(x_e, y_e, u_e)$ tel que

$$
\begin{cases}
x_e(1 - y_e) = 0 \\
-y_e(u_e - x_e) = 0 \\
\end{cases}
$$

- $x_e = 0$, impossible car $x(t) > 0$ pour tout $t \in [0, T]$
- $(1 - y_e) = 0 \Rightarrow y_e = 1$

Donc, $-y_e(u_e - x_e) = 0 \Rightarrow u_e = x_e$
Ainsi les points d'équilibre sont $(x_e, y_e, u_e) = (x_e, 1, x_e)$ avec $x_e > 0$


```python
x_e = np.linspace(0.1, 2, 100)
y_e = np.ones_like(x_e)  # y_e = 1
u_e = x_e  # u_e = x_e

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_e, y_e, label="$y_e = 1$", color="blue")
plt.plot(x_e, u_e, label="$u_e = x_e$", color="orange")
plt.title("Equilibrium Points $(x_e, y_e, u_e)$")
plt.xlabel("$x_e$")
plt.ylabel("Equilibrium Values")
plt.xlim(0, 2)
plt.ylim(0, 2)
plt.axhline(1, color="gray", linestyle="--", linewidth=0.7)  # y=1 line
plt.axvline(0, color="gray", linestyle="--", linewidth=0.7)  # x=0 line
plt.legend()
plt.show()
```

```python
def system(u, x, y):
    dxdt = x * (1 - y)
    dydt = -y * (u - x)
    return dxdt, dydt


# Create a grid of points
x_range = np.linspace(0, 5, 20)
y_range = np.linspace(0, 5, 20)
X, Y = np.meshgrid(x_range, y_range)

# Calculate the direction field for u(t) = 0
U1 = np.zeros_like(X)
V1 = np.zeros_like(Y)

for i in range(len(x_range)):
    for j in range(len(y_range)):
        U1[i, j], V1[i, j] = system(0, X[i, j], Y[i, j])

# Calculate the direction field for u(t) = 2
U2 = np.zeros_like(X)
V2 = np.zeros_like(Y)

for i in range(len(x_range)):
    for j in range(len(y_range)):
        U2[i, j], V2[i, j] = system(2, X[i, j], Y[i, j])

# Plotting the phase portraits
plt.figure(figsize=(14, 6))

# Without control
plt.subplot(1, 2, 1)
plt.quiver(X, Y, U1, V1, color="blue")
plt.plot(1, 4, "ko", markersize=8)  # IC
plt.title("Phase Portrait (u(t) = 0)")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("y")

# With control
plt.subplot(1, 2, 2)
plt.quiver(X, Y, U2, V2, color="red")
plt.plot(1, 4, "ko", markersize=8)
plt.title("Phase Portrait (u(t) = 2)")
plt.xlim(0, 5)
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("y")

plt.tight_layout()
plt.show()
```

- Pour $u(t) = 0$,  
  Il n'y a pas de contrôle sur les prédateurs. Les vecteurs ont tendance à être colinéaires et à aller vers $(x,y)=(0, \infty)$ jusqu'à l'extinction des nuisibles, puis rapidement le champ change vers $(0,0)$ quand les prédateurs meurent de faim faute de proies. Les points d'équilibre, quand le champ est nul, sont atteints lorsqu'il y a peu de proies et peu de prédateurs. Avec notre condition initiale (point noir), on arrive rapidement à $(0,0)$.

- Pour $u(t) = 2$,  
  On contrôle cette fois la prédation. Si on a trop de prédateurs, on diminue leur population, ce qui entraîne la population de nuisibles à croître doucement, jusqu'à ce que cela entraîne la population de prédateurs à croître elle aussi, et ainsi de suite. On voit bien le pattern circulaire sur le graphe. Les points d'équilibre sont plus nombreux, et leurs gradients sont nuls.


```python
m = GEKKO()
nt = 501
m.time = np.linspace(0, 1, nt)

a = 3

# Variables
x = m.Var(value=1)
y = m.Var(value=4)
p = np.zeros(nt)
p[-1] = 1
final = m.Param(value=p)

# Optimize final time
tf = m.FV(value=1.0, lb=0.1, ub=100.0)
tf.STATUS = 1

# Control
u = m.MV(value=0, lb=1, ub=3)
u.STATUS = 1

# Equations
m.Equation(x.dt() == x * tf * (1 - y * tf))
m.Equation(y.dt() == -y * tf * (u * tf - x * tf))

# Terminal constraints
terminal_penalty_x = 1e15 * (x * final - a) ** 2  # x(T) = a
terminal_penalty_y = 1e15 * (y * final - 1) ** 2  # y(T) = 1
terminal_penalty = terminal_penalty_x + terminal_penalty_y


m.Obj(tf + terminal_penalty)

# Solve
m.options.IMODE = 6
m.solve(disp=False)

print("Final Time: ", tf.value[0])
print("a: ", a)
print("x(T): ", x.value[-1])
print("y(T): ", y.value[-1])

tm = np.linspace(0, tf.value[0], nt)

plt.figure(figsize=(12, 6))

plt.plot(tm, x.value, "k-", label=r"$x$ : nuisance insect")
plt.plot(tm, y.value, "b-", label=r"$y$ : predator insect")
plt.plot(tm, u.value, "r--", label=r"$u$ : control input")
plt.title("States and Control Input vs Time")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend(loc="best")

plt.tight_layout()
plt.show()
```

Il n'y qu'avec $a=3$ que le systeme atteind une solution satisfaisant $x(T)=a$ et $y(T)=1$.  
On voit qu'avec nos condition initiales les predateurs sont "controlé" drastriquement, la population de proies augement presque linéairement jusqu'à l'introdution de nouveaux prédateurs, la population de proies diminue alors rapidement pour atteindre la contraite $x(T)=a$ tout en aillant $y(T)=1$.


```python
m = GEKKO()
nt = 501
m.time = np.linspace(0, 1, nt)

a = 4

# Variables
x = m.Var(value=1)
y = m.Var(value=4)
p = np.zeros(nt)
p[-1] = 1
final = m.Param(value=p)
z = m.Var(value=0)

# Optimize final time
tf = m.FV(value=1.0, lb=0.1, ub=100.0)
tf.STATUS = 1

# Control
u = m.MV(value=0, lb=1, ub=3)
u.STATUS = 1

# Equations
m.Equation(x.dt() == x * tf * (1 - y * tf))
m.Equation(y.dt() == -y * tf * (u * tf - x * tf))
m.Equation(z.dt() == (u * tf) ** 2)

# Terminal constraints
terminal_penalty_x = (x * final - a) ** 2  # x(T) = a
terminal_penalty_y = (y * final - 1) ** 2  # y(T) = 1
terminal_penalty = 1e15 * (terminal_penalty_x + terminal_penalty_y)


m.Obj(tf + z * final + terminal_penalty)

# Solve
m.options.IMODE = 6
m.solve(disp=False)

print("Final Time: ", tf.value[0])
print("a: ", a)
print("x(T): ", x.value[-1])
print("y(T): ", y.value[-1])

tm = np.linspace(0, tf.value[0], nt)

plt.figure(figsize=(12, 6))

plt.plot(tm, x.value, "k-", label=r"$x$ : nuisance insect")
plt.plot(tm, y.value, "b-", label=r"$y$ : predator insect")
plt.plot(tm, u.value, "r--", label=r"$u$ : control input")
plt.title("States and Control Input vs Time")
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend(loc="best")

plt.tight_layout()
plt.show()
```
