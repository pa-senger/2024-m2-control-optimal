# TP 3 : Controle optimal de l'équation de la chaleur


On suppose que $y_u$ est la solution de l'équation de la chaleur suivante :

$$
\begin{cases}
\partial_t y_u(t,x) - \partial_{xx} y_u(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
y_u(0,x) = 0 & \quad \text{pour } x \in [0,L]\\
y_u(t,0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x y_u(t,L) = u(t) &\quad \text{pour } t \in [0,T] \text{ Neumann non homogène}
\end{cases}
$$

On cherche à minimiser la fonctionnelle coût suivante :

$$
inf_{u \in L^2(0,T)} J(u) = \frac{1}{2} \int_0^T (y_u(t,L)-z_d(t))^2 + \frac{\epsilon}{2} \int_0^T u(t)^2
$$

où $\epsilon > 0$ est un paramètre fixé et $z_d \in L^2(0,T)$ est donnée.


### 1. Montrons que J est différentiable

Montrons que l'application $u \in L^2(0,T) \mapsto y_u \in H_0^1(0, T)$ est différentiable.  
Posons $z_h = \frac{y_{u+h} - y_u}{h}$, alors $z_h$ est solution de l'équation de la chaleur suivante :
$$
\begin{cases}
\partial_t z_h(t,x) - \partial_{xx} z_h(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
z_h(0,x) = 0 & \quad \text{pour } x \in [0,L]\\
z_h(t,0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x z_h(t,L) = \frac{u(t+h) - u(t)}{h} &\quad \text{pour } t \in [0,T] 
\end{cases}
$$

En passant à la limite $h \to 0$, on obtient:
$$
\begin{cases}
\partial_t z_h(t,x) - \partial_{xx} z_h(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
z_h(0,x) = 0 & \quad \text{pour } x \in [0,L]\\
z_h(t,0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x z_h(t,L) = v(t) &\quad \text{pour } t \in [0,T]
\end{cases}
$$

Si $u \mapsto y_u$ est différentiable, alors sa différentielle est $h \mapsto z_h$.  

Posons $S_h = y_{u+h} - y_h - hz_h$, alors $S_h$ est solution de l'équation de la chaleur suivante :
$$
\begin{cases}
\partial_t S_h(t,x) - \partial_{xx} S_h(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
S_h(0,x) = 0 & \quad \text{pour } x \in [0,L]\\
S_h(t,0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x S_h(t,L) = 0 &\quad \text{pour } t \in [0,T] 
\end{cases}
$$

Cette équation est homogène, les conditions initiales et aux limites sont nulles, donc par unicité de la solution, 
$$S_h = 0 \in o(||h||)$$  

Montrons que $h \mapsto z_h$ est linéaire.  
Soit $S = z_{h + \lambda k } - z_h - \lambda z_k$, on remarque que $S$ est solution de l'équation de la chaleur suivante :
$$
\begin{cases}
\partial_t S(t,x) - \partial_{xx} S(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
S(0,x) = 0 & \quad \text{pour } x \in [0,L]\\
S(t,0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x S(t,L) = 0 &\quad \text{pour } t \in [0,T]
\end{cases}
$$

Cette équation est homogène, les conditions initiales et aux limites sont nulles, donc par unicité de la solution,
$$S = 0 $$
Donc $z_{h + \lambda k } - z_h - \lambda z_k = 0$.  
Donc $h \mapsto z_h$ est linéaire.

Montrons que $h \mapsto z_h$ est continue.  
En multipliant l'équation de $z_h$ par $z_h$ et en intégrant sur $[0,T] \times [0,L]$, on obtient :  
$$\int_0^T \int_0^L z_h\partial_t z_h  -  z_h\partial_{xx} z_h = 0$$

Par intégration par parties, on obtient que :
$$\int_0^T \int_0^L | \nabla z_h |^2  + \int_0^L \int_0^T z_h^2 = \int_0^T z_h(t,L)h(t)$$

Par Cauchy-Schwarz, on obtient que :
$$\int_0^T z_h(t,L)h(t) \leq ||z_h||_{L^2(0,T)}||h||_{L^2(0,T)}$$

Donc, $||z_h||_{L^2(0,T)} \leq ||h||_{L^2(0,T)}$  
Donc $h \mapsto z_h$ est continue.  
DOnc $u \mapsto y_u$ est différentiable de différentielle $h \mapsto z_h$.  
Par compositon d'application différentiable, J est différentiable.

Calcul du gradient :  
On introduit $p$ la solution de l'équation adjointe suivante :

$$
\begin{cases}
\partial_t p(t,x) - \partial_{xx} p(t,x) = 0 &\quad \text{pour } (t,x) \in (0,T) \times (0,L)\\
p(T,x) = 0 & \quad \text{pour } x \in [0,L]\\
p(t, 0) = 0 &\quad \text{pour } t \in [0,T] \text{ Dirichlet homogène}\\
\partial_x p(t,L) = y_u(t,L) - z_d(t) &\quad \text{pour } t \in [0,T]
\end{cases}
$$

$$ J(u) = \int_0^T L(t, y_u(t,L), u(t)) dt$$


$$
\begin{align*}
 J(u + h) - J(u) &= \int_0^T L(t, y_{u+h}(t,L), u+h(t)) - L(t, y_u(t,L), u(t)) dt \\
&= \int_0^T \partial_{y_u} L(t, y_u(t), u) z_h(t) + \partial_u L(t, y_u(t), u) h(t) dt + o(||h||) 
\end{align*}
$$

On ulilise l'équation adjointe et des intégrations par parties pour obtenir :
$$
\nabla J(u) = \int_0^T \partial_{y_u} p(t,L) \partial_u y_u(t,L) dt
$$

### 2. Méthode numérique de type difference finie centrées en espace et décentrée en temps.

Algo :

1. On créer une grille decalée (Neumann)
2. On créer la matrice de Poisson
3. On initialise notre solution $U$ avec une fonction $u_0$.
4. On resoud le schema d'Euler explicite en temps pour trouver $U$.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
```

```python
def neumann_discretisation(N, L):
    dx = L / N 
    x = np.array([(i - 0.5) * dx for i in range(1, N + 1)])
    return x


x = neumann_discretisation(2, 1)
x
```

```python
def poisson_matrix(N):
    A = np.zeros((N, N))
    for i in range(1, N + 1):
        A[i - 1, i - 1] = 2
        if i < N:
            A[i - 1, i] = -1
            A[i, i - 1] = -1
    A[N - 1, N - 1] = 1
    return A


A = poisson_matrix(5)
A
```

```python
def u_boundary(t):
    return 1


def simulate_explicite_euler(u0, Nx, T, dt, L):
    x = neumann_discretisation(Nx, L)
    dx = L / Nx
    Nt = int(T / dt)
    U = np.zeros(Nx)
    U_new = np.zeros(Nx)
    A = poisson_matrix(Nx)
    cfl = dt / dx**2
    print(f"CFL = {cfl}")

    # Initialisation
    for i, xi in enumerate(x):
        U[i] = u0(xi)

    B = np.eye(Nx) - dt / dx**2 * A
    # B = np.eye(Nx) + dt / dx**2 * A
    bk = np.zeros(Nx)

    for t in range(1, Nt + 1):
        tn = t * dt
        
        bk[-1] = u_boundary(tn)
        U_new = np.dot(B, U) + dt/dx * bk
        # U_new = sc.gmres(B, U + dt/dx * bk)[0]
        # U_new = np.linalg.solve(B, U + dt/dx * bk)

        U = U_new.copy()

    return x, U
```

On introduit $ q(t, x)$ définie par :

$$
q(t, x) = p(T - t, x)
$$

$ q(t, x) $ représente l'état adjoint évalué à partir du temps final $ T $, et elle satisfait les conditions suivantes :

1. **Dérivées de $ q(t,x) $** :  
   $$
   \frac{\partial q}{\partial t}(t, x) = -\frac{\partial p}{\partial t}(T - t, x)
   $$
   $$
   \frac{\partial^2 q}{\partial x^2}(t, x) = \frac{\partial^2 p}{\partial x^2}(T - t, x)
   $$

2. **Équation pour $ q(t,x) $** :  
   En utilisant les relations ci-dessus, on obtient l'équation suivante pour $ q(t, x) $ :
   $$
   \frac{\partial q}{\partial t}(t, x) - \frac{\partial^2 q}{\partial x^2}(t, x) = 0
   $$
   Donc $ q(t,x) $ satisfait la même équation de diffusion que $ p(t,x) $, mais avec le temps inversé.

3. **Conditions aux limites pour $ q(t,x) $** :  
   Les conditions aux limites pour $ q(t,x) $ sont les suivantes :
   $$
   q(0, x) = 0 \quad \text{pour } x \in [0,L]
   $$
   $$
   q(t, 0) = 0 \quad \text{pour } t \in [0,T]
   $$
   $$
   \frac{\partial q}{\partial x}(t, L) = y_u(T - t, L) - z_d(T - t) \quad \text{pour } t \in [0,T]
   $$

Méthode numérique pour le calcul du gradient de $ J(u) $:

- Calcul de la solution $y_u(t,x)$ :  
Résoudre l'équation de la chaleur numériquement pour un contrôle $u$ donné.

- Calcul de $p(t,x)$ : 
Résoudre l'équation adjointe pour en utilisant des conditions aux limites appropriées.

- Calcul de $q(t,x)$ : 
Calculer $q(t,x)=p(T−t,x)$, ce qui est équivalent à résoudre l'équation de la chaleur avec un temps inversé pour $p$


```python
def compute_adjoint(x, L, T, dt, y_u, z_d):
    # Initialisation de p(t, x)
    Nx = len(x)
    p = np.zeros(Nx)

    # Résolution du problème adjointe à l'aide d'un schéma rétrograde
    # On parcourt les temps dans l'ordre décroissant
    for t in range(T - 1, -1, -1):  # On résout dans l'ordre rétrograde
        # Conditions aux limites sur p
        p[0] = 0  # Dirichlet homogène à x = 0
        p[-1] = y_u[t] - z_d[t]  # Condition de Neumann à x = L

        A = poisson_matrix(Nx)
        B = np.eye(Nx) - dt / (L / Nx) ** 2 * A
        p_new = np.dot(B, p)

        p = p_new.copy()

    return p


# Calcul du gradient de la fonction coût
def compute_gradient(u, x, L, T, epsilon, dt, z_d):
    gradient = np.zeros_like(u)

    # Résolution du problème direct pour y_u(t,x)
    _, y_u = simulate_explicite_euler(
        lambda x: u[int(x / L * len(u))], len(u), T, dt, L
    )

    # Résolution du problème adjoint pour p(t,x)
    p = compute_adjoint(x, L, T, dt, y_u, z_d)

    # Calcul du gradient
    for i in range(len(u)):
        delta_u = np.zeros_like(u)
        delta_u[i] = 1e-5  # Variation dans u[i]

        # Résoudre avec u + delta_u
        _, y_u_delta = simulate_explicite_euler(
            lambda x: u[int(x / L * len(u))] + delta_u[int(x / L * len(u))],
            len(u),
            T,
            dt,
            L,
        )

        # Calcul de la variation du coût
        J1 = 0.5 * np.sum((y_u - z_d) ** 2) + 0.5 * epsilon * np.sum(u**2)
        J2 = 0.5 * np.sum((y_u_delta - z_d) ** 2) + 0.5 * epsilon * np.sum(
            (u + delta_u) ** 2
        )

        # Différentielle du coût par rapport à u[i]
        gradient[i] = (J2 - J1) / 1e-5  # Approximation par différences finies
    return gradient


# Fonction de coût pour l'optimisation
def objective(u, L, T, epsilon, dt, z_d):
    _, y_u = simulate_explicite_euler(
        lambda x: u[int(x / L * len(u))], len(u), T, dt, L
    )
    return 0.5 * np.sum((y_u - z_d) ** 2) + 0.5 * epsilon * np.sum(u**2)


# Algorithme de gradient
def gradient_descent(
    u_initial, L, T, epsilon, dt, z_d, max_iter=100, learning_rate=0.01
):
    u = u_initial.copy()
    for iter in range(max_iter):
        gradient = compute_gradient(u, x, L, T, epsilon, dt, z_d)
        u = u - learning_rate * gradient  # Mise à jour du contrôle
        if iter % 10 == 0:
            print(f"Iteration {iter}, Coût : {objective(u, L, T, epsilon, dt, z_d)}")
    return u
```

```python
# Paramètres
L = 5
T = 1
epsilon = 0.1
Nx = 50
dt = 0.005
z_d = np.sin(np.linspace(0, T, Nx))  # Cible de contrôle

# Initialisation 
u_initial = np.zeros(Nx)

# Résolution par gradient
u_optimal = gradient_descent(u_initial, L, T, epsilon, dt, z_d)

plt.plot(u_optimal, label="Contrôle optimal")
plt.legend()
plt.show()
```
